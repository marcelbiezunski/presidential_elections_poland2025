```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('protokoly_obwody_1.csv', sep=";", quotechar='"', encoding="utf-8", engine="python")
df2 = pd.read_csv('protokoly_obwody_2.csv', sep=";", quotechar='"', encoding="utf-8", engine="python")
```

```python
df1.columns.tolist()
```

```python
df2.columns.tolist()
```

```python
df1 = df1[["Nr komisji", "Teryt Gminy", 'Teryt Powiatu', "Liczba głosów ważnych oddanych łącznie na wszystkich kandydatów", "NAWROCKI Karol Tadeusz", "TRZASKOWSKI Rafał Kazimierz"]]
df1['suma głosów Nawrockiego i Trzaskowskiego'] = (df1["NAWROCKI Karol Tadeusz"] + df1["TRZASKOWSKI Rafał Kazimierz"])
```

```python
df1
```

```python
df2 = df2[["Nr komisji", "Teryt Gminy", 'Teryt Powiatu', "Liczba głosów ważnych oddanych łącznie na obu kandydatów", "NAWROCKI Karol Tadeusz", "TRZASKOWSKI Rafał Kazimierz"]]
```

```python
df2
```

```python
merged = pd.merge(df1, df2, on=["Nr komisji", 'Teryt Gminy', 'Teryt Powiatu'], suffixes=("1_tura", "2_tura"))
```

```python
merged
```

```python
merged['procentowo_1_tura'] = merged['NAWROCKI Karol Tadeusz1_tura'] / merged['suma głosów Nawrockiego i Trzaskowskiego'] *100
merged['procentowo_2_tura'] = merged['NAWROCKI Karol Tadeusz2_tura'] / merged['Liczba głosów ważnych oddanych łącznie na obu kandydatów'] *100
```

```python
merged
```

```python
filtered = merged[merged["Liczba głosów ważnych oddanych łącznie na obu kandydatów"] >= 500]
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(filtered['procentowo_1_tura'], filtered['procentowo_2_tura'], alpha=0.7)
plt.xlabel('Procent głosów Nawrockiego względem Trzaskowskiego (1. tura)')
plt.ylabel('Procent głosów Nawrockiego względem Trzaskowskiego (2. tura)')
plt.title('Porównanie poparcia Nawrockiego względem Trzaskowskiego (1 oraz 2 tura)')
plt.grid(True)
plt.plot([0, 100], [0, 100], 'r--', label="Linia równości")
plt.legend()
plt.show()
```
